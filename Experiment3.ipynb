```python
result = ("Maths",86,"A")
print("Subject:",result[0])
print("Marks:",result[1])
print("Grade:",result[2])
#unpacking a tuple
subject,marks,grade = result
print("subject:",subject)
print("marks:",marks)
print("grade:",grade)
#check for distinction
if marks>=75:
  print("Distinction")
```
